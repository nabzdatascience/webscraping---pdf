```python
#Import the following libraries
import pandas as pd # for data analysis and manipulation
import requests # for establishing network connection
from bs4 import BeautifulSoup # for extracting content from HTML or XML docs
```

```python
#include this url to scrape data from this website
url="https://parsehub.com/sandbox/showtimes"
```

```python
# Establishing connection to fetch the page
response=requests.get(url)
print(response)
```

```python
print(response.text)
```

```python
# Just execute the following print statement once.Since the application is working properly, comment out the print statement by adding a '#' before it so that next time it should not get executed
#print(response)
```

```python
# to parse the content, we will use BeautifulSoup library and the python's default libary,html.parser
soup=BeautifulSoup(response.text,"html.parser")
```

```python
print(soup)
```

```python
#prettify() is a function which allows you to easily view and locate specific data points within the HTML or XML page.
print(soup.prettify()) 
```

```python
# Create lists to store data temporarily
movie_name=[]
link=[]
movie_time=[]
```

```python
# Extracting the movie Titles and Links and storing the scraped data in the lists
for m in soup.find_all("a",class_="title"):
    movie_name.append(m.text)
    link.append(m.get("href"))
print(movie_name)
print(link)

   
```

```python
#Extract the movie time and store in the list
first_time=soup.find("span",class_="borderbox showtime imax first").text# first movie time
movie_time.append(first_time)
for t in soup.find_all("span",class_="borderbox showtime imax other"): # the remaining movie time
    movie_time.append(t.text)
print(movie_time)    
    
    
            
```

```python
#Dictionary to store the lists
dictionary={
 'Title':movie_name,
 
 'Link':link,
 'Time':movie_time,
 #'ImageLink':img_url   
    
}
```

```python
# store the dictionary in a dataframe using the pandas library and specify the names of the columns
df=pd.DataFrame(dictionary,columns=['Title','Link','Time'])
```
